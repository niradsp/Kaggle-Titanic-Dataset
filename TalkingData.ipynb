```python
def trainingset(train,ratio=0.8):
    ratio=len(train)*ratio
    ratio=int(ratio)
    X_train=train[:ratio]
    X_test=train[ratio:]
    return X_train,X_test
```

```python
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'auc',
    #'feature_fraction':1,
    'subsample_freq':1,
    'min_child_samples':100,
    'subsample':0.7,
    'colsample_bytree':0.9,
    'min_child_weight':0,
    'max_depth':3,
    'max_bin':100,
    'scale_pos_weight': 200,
    'num_leaves': 7,
    'learning_rate': 0.1,
    'verbose': 0,
    'nthread':10
}
```

```python
categorical=['app','device','os','channel','hour']#,'app_device','appdeviceos','ipappdeviceoschannel','ipappdeviceos','deviceoschannel']
predictors=list(X_train1.columns)
print("Training")
lgb_train=lgb.Dataset(X_train1.values,label=Y_train.values,categorical_feature=categorical,feature_name=predictors)
lgb_eval=lgb.Dataset(X_test1.values,label=Y_test.values,feature_name=predictors,categorical_feature=categorical)
gbm = lgb.train(params=params,train_set=lgb_train,valid_sets=[lgb_eval],categorical_feature=categorical,num_boost_round=1500,early_stopping_rounds=50)
```
